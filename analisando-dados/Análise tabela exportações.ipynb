### Puxar base de dados

```python
import pandas as pd
analise_inicial = pd.read_csv('exportacoes_franca.csv')
display(analise_inicial)
```

### Informações Gerais

```python
print(analise_inicial.info())
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
exportacao_ano = analise_inicial.groupby("Year").sum()
exportacao_ano = exportacao_ano[["US$ FOB"]]

def formatar(valor):
    return f"${valor:,.2f}"

exportacao_ano["US$ FOB"] = exportacao_ano["US$ FOB"].map(formatar)
display(exportacao_ano)
#exportacao_ano.plot(kind='bar')

```

### Quais os produtos mais exportados ao longo de todo o período?

```python
exportacao_produto = analise_inicial.groupby("SH4 Description").sum()
exportacao_produto = exportacao_produto[['US$ FOB']]
exportacao_produto = exportacao_produto.sort_values('US$ FOB', ascending=False)
exportacao_produto["US$ FOB"] = exportacao_produto["US$ FOB"].map(formatar)
display(exportacao_produto)
```

### Em 2020 qual cidade mais exportou para a França?

```python
tabela_2020 = analise_inicial.loc[analise_inicial["Year"]==2020, :]
exportacao_cidades = tabela_2020.groupby("City").sum()
exportacao_cidades = exportacao_cidades[['US$ FOB']]
exportacao_cidades = exportacao_cidades.sort_values('US$ FOB', ascending=False)
exportacao_cidades["US$ FOB"] = exportacao_cidades['US$ FOB'].map(formatar)
display(exportacao_cidades)
```
